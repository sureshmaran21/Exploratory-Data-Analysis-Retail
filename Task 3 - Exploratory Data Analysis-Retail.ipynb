# Visualizing Plots

```python
plt.figure(figsize=(16,18))
plt.bar('Sub-Category','Category',data=df1)
plt.title('Category va Sub Category')
plt.xlabel('Sub-Category')
plt.ylabel('Category')
plt.xticks(rotation=45)
plt.show()
```

```python
df1.hist(bins=50,figsize=(20,15))
plt.show();
```

```python
sns.scatterplot(x=df1['Ship Mode'],y=df1['Sales'],hue=df1['Category'])
```

```python
ds=df1.groupby('Category')['Profit','Sales'].agg('sum')
print(ds)
ds.plot.bar()
plt.legend(loc='upper left')
plt.title('Category wise Profit and Sale')
```

```python
df1['State'].value_counts()
```

```python
plt.figure(figsize=(15,15))
sns.countplot(x=df1['State'])
plt.xticks(rotation=90)
plt.title("State")
plt.show()
```

```python
from plotnine import *
Profit_plot=(ggplot(df1,aes(x='Sub-Category', y='Profit', fill='Sub-Category'))+geom_col()+coord_flip()+scale_fill_brewer(type='div',palette="Spectral")+theme_classic()+ggtitle('Pie Chart'))
display(Profit_plot)
```

Above Pie chart Shows the profit and loss of each and every subcategories. Here from the graph we can visualize that "binders" sub-category has suffered the highest amount loss and also profit amongst all other Sub-categories (For now we can't sat that what is the reason it may be because of discounts given on blinders subcategory) Next, "Copies" Sub- category has gain highest amount of profit with no loss. There are other sub-categories too who are not faced any kind of losses but their profit margins are also low
